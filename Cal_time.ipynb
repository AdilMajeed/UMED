# Conversion from Vidoes to Audio

```python
import os
from moviepy.editor import *

# Directory containing video files
video_directory = 'F:/thesis/temp/F4/F41'

# Directory where you want to save audio files
output_directory = 'F:/thesis/temp/F4/F41/Audio'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Get a list of all video files in the input directory
video_files = [file for file in os.listdir(video_directory) if file.endswith(".mp4")]

for video_file in video_files:
    # Construct the full path of the input video
    video_path = os.path.join(video_directory, video_file)

    # Load the video file
    video_clip = VideoFileClip(video_path)

    # Extract audio from the video
    audio_clip = video_clip.audio

    # Specify the output audio file path
    output_audio_path = os.path.join(output_directory, f"{os.path.splitext(video_file)[0]}.mp3")

    # Write the audio to the output file
    audio_clip.write_audiofile(output_audio_path)

    # Close the video and audio clips
    video_clip.close()
    audio_clip.close()

    print(f"Audio extracted and saved as {output_audio_path}")
```

# Converion from audio to text

```python
import os
import whisper
import csv
import openpyxl

# Initialize the Whisper ASR recognizer with the desired model
model = whisper.load_model("large")

# Directory containing your audio files
audio_directory = "F:/thesis/Data/F6/F71/Audio"  # Update this to your audio files directory
#audio_directory = "F:/Test"
# Function to transcribe an audio file
def transcribe_audio(audio_file):
    try:
        result = model.transcribe(audio_file, language="ur")
       # result = model.transcribe(filepath, language="pt", fp16=False, verbose=True)
        return result["text"]
    except Exception as e:
        return str(e)

# Create and open an Excel file for writing
output_file = "F:/thesis/Data/F6/Text/transcriptions_F71.xlsx"
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet.title = "Transcriptions"

# Write the header row to the Excel file
header = ["Audio File", "Label", "Transcription"]
sheet.append(header)
count = 0;
# Process all audio files in the directory and its subdirectories
for root, dirs, files in os.walk(audio_directory):
    for file in files:
        if file.endswith(".mp3"):
            audio_file = os.path.join(root, file)
            transcription = transcribe_audio(audio_file)
            
            print(transcription)
            label = file[0]
            count = count + 1
            print(count)
            # Add the data to the Excel sheet
            data = [audio_file, label, transcription]
            sheet.append(data)

# Save the Excel file
workbook.save(output_file)

print(f"Transcriptions saved to {output_file}")
```

```python
import os
from moviepy.editor import VideoFileClip

folder_path = 'E:/Study/VidCropper v 2.2/VidCropper/cropped/temp/'
#folder_path = 'F:/thesis/Data/New Data/Anger'

# Function to calculate total time in seconds
def calculate_total_time(folder_path):
    total_time = 0
    for filename in os.listdir(folder_path):
        if filename.endswith(".mp4"):
            video_path = os.path.join(folder_path, filename)
            video_clip = VideoFileClip(video_path)
            total_time += video_clip.duration
            video_clip.close()
    return total_time

total_time_seconds = calculate_total_time(folder_path)

# Convert total time to hours, minutes, and seconds
total_hours = int(total_time_seconds // 3600)
total_minutes = int((total_time_seconds % 3600) // 60)
total_seconds = int(total_time_seconds % 60)

print(f"Total time: {total_hours} hours, {total_minutes} minutes, {total_seconds} seconds")
```
